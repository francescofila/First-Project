Regressione Ridge in Pyhton

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import Ridge, RidgeCV, LinearRegression
```

```python
data  = fetch_california_housing(as_frame=True)
X = data.data
y = data.target

display(X)
```

```python
display (y)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_scaled, y)
y_pred_lin = lin_reg.predict(X_scaled)

mse_lin = mean_squared_error(y, y_pred_lin)
r2_lin = r2_score(y, y_pred_lin)

print(f"MSE (linear):\t{mse_lin:.4}\nR2 (Linear):\t{r2_lin:.4f}")
```

```python
r2_ridge = []
mse_ridge = []
alphas = np.logspace(-3, 3, 100)

for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(X_scaled, y)
    y_pred = ridge.predict(X_scaled)

    mse_ridge.append(mean_squared_error(y, y_pred))
    r2_ridge.append(r2_score(y, y_pred))

results = pd.DataFrame({
    "alpha":alphas,
    "MSE":mse_ridge,
    "R2": r2_ridge
})

display(results)
```

```python
alphas_cv = alphas
ridge_cv = RidgeCV(alphas=alphas, store_cv_results=True)

ridge_cv.fit(X_scaled, y)

print(f"Alpha RidgeCV:\t{ridge_cv.alpha_:.4f}")

y_pred_cv = ridge_cv.predict(X_scaled)
mse_cv = mean_squared_error(y, y_pred_cv)
r2_cv =  r2_score(y, y_pred_cv)

print(f"MSE:\t{mse_cv:.4f}\nR2:\t{r2_cv:.4f}")


```

```python
best_idx_man = np.argmin(mse_ridge)
best_alpha_man = alphas[best_idx_man]
best_mse_man = mse_ridge[best_idx_man]
best_r2_man = r2_ridge[best_idx_man]

print(f"Alpha Ridge:\t{best_alpha_man:.4f}\nMSE:\t{best_mse_man:.4f}\nR2\t{best_r2_man:.4f}")
```

```python
fig, ax1 = plt.subplots(figsize=(8, 6))

ax1.plot(alphas, mse_ridge, '-', color='firebrick')
ax1.set_xscale('log')
ax1.set_xlabel("Alphas")
ax1.set_ylabel("MSE", color='firebrick')

ax1.scatter(best_alpha_man, best_mse_man, s=120, zorder=5,
            label=f"best alpha manuale = {best_alpha_man:.4f}")
ax1.axvline(best_alpha_man, linestyle=":", linewidth=1)

ax1.scatter(ridge_cv.alpha_, mse_cv, marker="X", s=180, zorder=5,
            label=f"alpha RidgeCV = {ridge_cv.alpha_:.4f}")
ax1.axvline(ridge_cv.alpha_, linestyle="--", linewidth=1)


ax2 = ax1.twinx()
ax2.plot(alphas, r2_ridge, '--', label="R² Ridge", color = 'darkblue')
ax2.set_ylabel("R2", color='darkblue')

plt.title("Confronto MSE vs R2")
plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()


```
