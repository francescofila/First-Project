A — Import delle librerie

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import learning_curve, GridSearchCV
from sklearn.model_selection import cross_val_score, StratifiedKFold, train_test_split

from pandas.plotting import scatter_matrix
```

B — Caricamento del dataset

```python
data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='class')

df = X.copy()
df['class'] = y
display(df)
```

C — Statistiche descrittive e distribuzione classi

```python
print(df.describe())
print(df.groupby('class').size())
```

D — Visualizzazioni iniziali

```python
# Boxplot

"""
Serve a vedere:

- distribuzione
- mediana
- dispersione
- possibili outlier 
"""

df.boxplot(figsize=(8, 6), grid=False)
plt.show()
```

```python
# Istogrammi 
"""
come sono distribuiti i valori di ogni variabile
"""

df.hist(figsize=(10, 8))
plt.show()
```

```python
# Scatter matrix

"""
Serve a osservare:

- relazioni coppia per coppia tra le feature
- possibile separazione tra classi
- pattern visivi utili prima del training
"""
scatter_matrix(df, figsize=(10, 8))
plt.show()
```

E — Train/Test split


```python
"""
Cosa fa

Dividiamo il dataset:

80% training
20% test
"""


X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    shuffle=True,
    test_size=0.20,
    random_state=42,
)
```

F — Definizione dei modelli

Costruiamo una lista di modelli da confrontare.

Modelli
SVM: cerca frontiere di separazione ottimali
KNN: classifica in base ai vicini più prossimi
CART: albero decisionale
LR: regressione logistica per classificazione

```python
models = [
    ('SVM', SVC(gamma='auto')),
    ('KNN', KNeighborsClassifier()),
    ('CART', DecisionTreeClassifier()),
    ('LR', LogisticRegression(max_iter=1000))
]
```

G - Cross-validation sui modelli

Per ogni modello:

creiamo una validazione incrociata stratificata a 10 fold
calcoliamo l’accuracy su ogni fold
salviamo i risultati
stampiamo media e deviazione standard

```python
names = []
results = []

K = 10
for name, model in models:
    kfold = StratifiedKFold(n_splits=K, shuffle=True, random_state=42)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='accuracy')
    names.append(name)
    results.append(cv_results)
    print(f"{name}:\n\tMEAN:\t{cv_results.mean():.4f}\n\tSTD: \t{cv_results.std():.4f}")
```

H — Boxplot di confronto

Mostra visivamente le distribuzioni delle accuracy ottenute dai modelli nei vari fold.

```python
plt.boxplot(results, tick_labels=names)
plt.title("Confronto Algoritmi")
plt.show()
```

I — Addestramento del modello scelto

```python
model = SVC(gamma='auto')
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

L — Valutazione finale del modello

```python
print(f"Accuracy:\t{accuracy_score(y_test, predictions)}")
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

M — Learning curve del modello base

```python
train_sizes, train_scores, val_scores = learning_curve(
    model,
    X_train,
    y_train,
    cv=5,
    n_jobs=-1, 
    scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10),
)


train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, "o-", color="red", label="Training score")
plt.fill_between(
    train_sizes,
    train_mean - train_std,
    train_mean + train_std,
    alpha=0.1,
    color="red"
)
plt.plot(train_sizes, val_mean, "o-", color="blue", label="Validation score")
plt.fill_between(
    train_sizes,
    val_mean - val_std,
    val_mean + val_std,
    alpha=0.1,
    color="blue"
)


plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.title("Learning Curves - SVC")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

N — Grid Search su SVC

```python
svc = SVC(gamma='auto')

param_grid = {
    'C': np.logspace(-2, 2, 5),
}


# stiamo ottimizzando solo l’iperparametro C


grid = GridSearchCV(
    cv=5,
    n_jobs=-1,
    estimator=svc,
    scoring='accuracy', 
    param_grid=param_grid
)

# Proviamo tutti i valori di C indicati e sceglie quello con accuracy media migliore in cross-validation
# E restituiamo -> miglior valore di C e -> miglior score medio


grid.fit(X_train, y_train)

print(f"C:\t{grid.best_params_['C']}")
print(f"CV:\t{grid.best_score_:.3f}")
```

```python
best_model = grid.best_estimator_

best_predictions = best_model.predict(X_test)

print("\nValutazione modello ottimizzato:")
print(f"Accuracy: {accuracy_score(y_test, best_predictions):.4f}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, best_predictions))

print("\nClassification Report:")
print(classification_report(y_test, best_predictions))



train_sizes, train_scores, val_scores = learning_curve(
    best_model,
    X_train,
    y_train,
    cv=5,
    n_jobs=-1,
    scoring="accuracy",
    train_sizes=np.linspace(0.1, 1.0, 10)
)

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, "o-", color="red", label="Training score")
plt.fill_between(
    train_sizes,
    train_mean - train_std,
    train_mean + train_std,
    alpha=0.1,
    color="red"
)

plt.plot(train_sizes, val_mean, "o-", color="blue", label="Validation score")
plt.fill_between(
    train_sizes,
    val_mean - val_std,
    val_mean + val_std,
    alpha=0.1,
    color="blue"
)

plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.title("Learning Curves - SVC ottimizzato")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

P — PCA finale e visualizzazione 2D

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

best_model.fit(X_pca, y)

plt.figure(figsize=(10, 8))

# Scatter dei punti ridotti in 2D
scatter = plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=y,
    cmap="coolwarm",
    edgecolors="k",
    s=100
)



x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

# Predizione su tutti i punti della griglia
Z = best_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Area colorata dei confini decisionali
plt.contourf(xx, yy, Z, cmap="coolwarm", alpha=0.3)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA 2D + Decision Boundary del miglior modello")
plt.legend(*scatter.legend_elements(), title="Classi")
plt.grid(True)
plt.show()
```
