Combinazione Supervised & unsupervised Learning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
```

Diabets Dataset

```python
data = load_diabetes()

print(data.DESCR)
```

```python
X, y = data.data, data.target
df = pd.DataFrame(X, columns=data.feature_names)
df ['target'] = y
df.dtypes
```

1.Linear Regression (base)

```python
model_base = LinearRegression()
model_base.fit(X,y)
y_pred_base = model_base.predict(X)

errors_base = np.abs(y -y_pred_base)
mae_base = np.mean(errors_base)

print(f"MODEL BASE\nerr: {errors_base}\nMAE: {mae_base}")
```

2. Linear regression (clustering)

```python
kmeans = KMeans(n_clusters=3, random_state=42)

clusters = kmeans.fit_predict(X)

X_new = np.c_[X, clusters]


model_clustering = LinearRegression()
model_clustering.fit(X_new, y)

y_pred_clustering = model_clustering.predict(X_new)

errors_clustering = np.abs(y-y_pred_clustering)
mae_clustering = np.mean(errors_clustering)
print(f"MODEL CLUSTERING\nerr:{errors_clustering}\nMAE:{mae_clustering}")
```

3. Errors PLot

```python
cumulative_base = np.cumsum((y - y_pred_base)**2) / (np.arange(len(y))+1)
cumulative_clustering = np.cumsum((y-y_pred_clustering)**2) / (np.arange(len(y))+1)

plt.figure(figsize=(10,6))

plt.plot(cumulative_base, label= "Cumulative Risk (base)")
plt.plot(cumulative_clustering, label="Cumulative Risk (clustering)")
plt.xlabel("# Dati osservati ")
plt.ylabel("Mean Sqared Error")
plt.title( "Confronto MSE")
plt.legend()

plt.show()

```

4. Linear Regression Plot

```python
plt.figure(figsize=(8, 6))

plt.scatter(y, y_pred_base, color='skyblue', edgecolors='k', label="Datapoint")
plt.plot([y.min(), y.max()], [y_pred_base.min(), y_pred_base.max()], color='red', linewidth=2, label= "Linear Regression Base")
plt.plot([y.min(), y.max()], [y_pred_clustering.min(), y_pred_clustering.max()], color='blue', linewidth=2, label= "Linear Regression Clustering")

plt.xlabel("valore Reale")
plt.ylabel("Valore Predetto")
plt.legend()
plt.title("Confronto Linear regression")

plt.show()
```
