```python
# %% [markdown]
# # Rand Index in Python
#
# In questo notebook:
# - carichiamo il dataset `digits`
# - osserviamo i dati in forma tabellare
# - applichiamo K-Means con `K = 10`
# - confrontiamo i cluster ottenuti con le etichette reali
# - valutiamo il risultato tramite Rand Index

# %%
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import rand_score
from sklearn.datasets import load_digits

# %% [markdown]
# ## 1. Caricamento del dataset
#
# Il dataset `digits` contiene immagini di cifre scritte a mano.
# Ogni osservazione è rappresentata da feature numeriche derivate dai pixel.
# La variabile `target` contiene la cifra reale associata a ciascuna osservazione.

# %%
data = load_digits()

X = data.data
y = data.target

print("Shape di X:", X.shape)
print("Shape di y:", y.shape)

# %% [markdown]
# ## 2. Creazione del DataFrame
#
# Convertiamo i dati in un DataFrame Pandas per ispezionarli meglio.
# In alcune versioni di scikit-learn `feature_names` potrebbe non essere disponibile:
# per questo costruiamo nomi di colonna robusti in modo esplicito.

# %%
feature_names = [f"pixel_{i}" for i in range(X.shape[1])]

df = pd.DataFrame(X, columns=feature_names)
df["target"] = y

print("Prime 5 righe del dataset:")
display(df.head())

print("\nDimensioni del DataFrame:")
print(df.shape)

# %% [markdown]
# ## 3. Clustering con K-Means
#
# Impostiamo `K = 10` perché nel dataset esistono 10 classi reali (da 0 a 9).
# Attenzione: K-Means non usa le etichette reali durante il training.
# Le useremo solo dopo, nella fase di valutazione.

# %%
K = 10

kmeans = KMeans(
    n_clusters=K,
    random_state=42,
    n_init=10
)

y_pred = kmeans.fit_predict(X)

print("Prime 20 etichette predette dal clustering:")
print(y_pred[:20])

# %% [markdown]
# ## 4. Calcolo del Rand Index
#
# Il Rand Index confronta la partizione reale (`y`) e quella stimata (`y_pred`)
# osservando la coerenza delle coppie di punti.
#
# - valore vicino a 1 -> clustering molto coerente con la struttura reale
# - valore vicino a 0 -> clustering poco coerente

# %%
ri = rand_score(y, y_pred)
print(f"Rand Index: {ri:.3f}")

# %% [markdown]
# ## 5. Interpretazione finale
#
# Il Rand Index non confronta direttamente il nome numerico dei cluster con il nome delle classi,
# perché nel clustering le etichette assegnate ai gruppi sono arbitrarie.
#
# Confronta invece la struttura:
# verifica se due punti che stanno insieme nella realtà restano insieme anche nel clustering,
# e se due punti che sono separati nella realtà restano separati anche nella soluzione stimata.

# %%
if ri > 0.90:
    interpretation = "Ottima coerenza tra clustering stimato e classi reali."
elif ri > 0.75:
    interpretation = "Buona coerenza: il clustering cattura gran parte della struttura reale."
elif ri > 0.50:
    interpretation = "Coerenza intermedia: il clustering coglie solo in parte la struttura reale."
else:
    interpretation = "Coerenza debole: il clustering non rappresenta bene le classi reali."

print("Interpretazione:", interpretation)
```

```python
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.preprocessing import StandardScaler

# Dataset
data = load_digits()
X = data.data
y_true = data.target

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Clustering
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
y_pred = kmeans.fit_predict(X_scaled)


# RI
ri = rand_score(y, y_pred)
print(f"Rand Index: {ri:.3f}")


# ARI
ari = adjusted_rand_score(y_true, y_pred)
print(f"Adjusted Rand Index: {ari:.4f}")
```
