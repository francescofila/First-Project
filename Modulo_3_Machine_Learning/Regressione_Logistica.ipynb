1 - improt librerie

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
```

2 - caricamento dati

```python
data = load_iris(as_frame=True)
df = data.frame

display(df.tail())

X = df[data.feature_names]
y = df["target"]

print(X.shape)
```

3 addestramento modello 

```python
model = LogisticRegression(max_iter=5000)
model.fit(X, y)
```

4 - predizioni

```python
y_pred = model.predict(X)
```

5 - accuratezza

```python
corrette = np.sum(y==y_pred)
totali=len(y)
acccuracy = (corrette / totali) * 100

print(f"Etichette corrette: {corrette}/{totali}")
print(f"accuracy (%): {acccuracy:.2f}%")
```

```python
# 6. Scatter plot reali vs predetti
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(range(totali), y, alpha=0.6, label="Reali")
plt.scatter(range(totali), y_pred, alpha=0.6, label="Predette")
plt.title("Grafico Reali-Predette")
plt.xlabel("Indice campione")
plt.ylabel("Classe")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

# 7. Scatter plot Petal Length vs classe


```python
plt.figure(figsize=(8, 6))
plt.scatter(df["petal length (cm)"], y, alpha=0.4, label="Reali")
plt.scatter(df["petal length (cm)"], y_pred, alpha=0.4, label="Predette")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Classe")
plt.title("Relazione Petal Length - Classe")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
