# Progetto 2 - Regressione Lineare, Ridge e LASSO


Import e Caricamenteo Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_linnerud
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score

data = load_linnerud(as_frame=True)

X = data.data.copy()
Y = data.target.copy()

print("Shape X:", X.shape)
print("Shape Y:", Y.shape)

display(X.head())
display(Y.head())
```

Analisi descrittiva iniziale

```python
print("=== Descrizione feature X ===")
display(X.describe())

print("=== Descrizione target Y ===")
display(Y.describe())
```

Standardizzazione delle feature

```python
scaler_X = StandardScaler()
X_std = scaler_X.fit_transform(X)

X_std_df = pd.DataFrame(X_std, columns=X.columns)

print("=== Feature standardizzate ===")
display(X_std_df.head())
```

Target engineering - PCA sui target

```python
scaler_Y = StandardScaler()
Y_std = scaler_Y.fit_transform(Y)

Y_std_df = pd.DataFrame(Y_std, columns=Y.columns)

print("=== Target standardizzati ===")
display(Y_std_df.head())

pca_target = PCA(n_components=1)
y_pca = pca_target.fit_transform(Y_std).ravel()

print("Explained variance ratio target PC1:", pca_target.explained_variance_ratio_[0])

target_loadings = pd.Series(
    pca_target.components_[0],
    index=Y.columns,
    name="PC1_target_loading"
)

display(target_loadings.to_frame())
```
