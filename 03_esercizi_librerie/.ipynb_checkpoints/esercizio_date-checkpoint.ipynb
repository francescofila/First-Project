# Esercizio — Serie temporale 30 giorni

Obiettivo: generare 30 date, valori casuali, DataFrame con indice temporale, grafico a linee.

Librerie: pandas, numpy, matplotlib


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
date = pd.date_range(start=pd.Timestamp.today().normalize(), periods=30, freq="D")
date
```

```python
np.random.seed(42)  # così i numeri sono ripetibili (utile per studiare)

valori = np.random.randint(0, 100, size=30)

df = pd.DataFrame({"valore": valori}, index=date)
df.index.name = "data"

df.head()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df.index, df["valore"], marker="o")
plt.title("Valori casuali su 30 giorni")
plt.xlabel("Data")
plt.ylabel("Valore")
plt.grid(True)
plt.tight_layout()
plt.show()
```
