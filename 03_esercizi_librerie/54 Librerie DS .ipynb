```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr = np.arange(1,11)

quadrati = arr**2
cubi = arr**3
```

```python
df = pd.DataFrame({

    "Numero": arr,
    "Quadrato": quadrati,
    "Cubo": cubi
    
})
```

```python
print(df)

# Grafico

plt.plot(df["Numero"], df["Quadrato"], label = "Quadrati")
plt.plot(df["Numero"], df["Cubo"], label = "Cubi")
plt.xlabel("Numero")
plt.ylabel("Valore")
plt.title("Quadrati e Cubi")
plt.legend()
plt.show()
```
