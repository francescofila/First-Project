# Esercizio — Array 1..10, quadrati e cubi

Obiettivo: NumPy (array/potenze) → Pandas (DataFrame) → Matplotlib (grafico linee).


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Array 1....10, + quadrati e cubi

numeri = np.arange(1, 11)

quadrati = numeri**2
cubi = numeri**3

numeri, quadrati, cubi
```

```python
# DataFrame

df= pd.DataFrame({
    "numero": numeri,
    "quadrato": quadrati,
    "cubo": cubi
})
```

```python
plt.figure(figsize=(10, 4))

plt.plot(df["numero"], df["numero"], marker="o", label="numero")
plt.plot(df["numero"], df["quadrato"], marker="o", label="quadrato")
plt.plot(df["numero"], df["cubo"], marker="o", label="cubo")

plt.title("Numeri, quadrati e cubi (1-10)")
plt.xlabel("Numero")
plt.ylabel("Valore")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
