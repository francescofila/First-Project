```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(42)
giorni = ["Lun", "Mar", "Mer", "Gio", "Ven"]
fasce = ["Mattina", "Pomeriggio", "Sera"]

df = pd.DataFrame({
    "giorno": np.random.choice(giorni, 300),
    "fascia": np.random.choice(fasce, 300),
})

mu = df["fascia"].map({"Mattina": 7, "Pomeriggio": 5.3, "Sera": 6.2}).to_numpy()
sigma = df["fascia"].map({"Mattina": 1.0, "Pomeriggio": 1.4, "Sera": 0.9}).to_numpy()
df["soddisfazione"] = np.clip(np.random.normal(mu, sigma), 0, 10)


# DEMO 1 — Boxplot personalizzato + punti

plt.figure(figsize=(8,5))
sns.boxplot(
    data=df, x="giorno", y="soddisfazione", hue="fascia",
    palette="Set2", linewidth=1.3, fliersize=6,
    showmeans=True,
    meanprops={"marker":"o","markerfacecolor":"black","markeredgecolor":"black"}
)
sns.stripplot(
    data=df, x="giorno", y="soddisfazione", hue="fascia",
    dodge=True, alpha=0.35, jitter=True, size=3, color="gray"
)
plt.title("Boxplot soddisfazione per giorno e fascia")
plt.legend(title="Fascia", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.show()



# DEMO 2 — Violinplot avanzato (quartili, cut)

plt.figure(figsize=(8,5))
sns.violinplot(
    data=df, x="giorno", y="soddisfazione", hue="fascia",
    inner="quartile", cut=0, palette="pastel"
)
plt.title("Violinplot avanzato (quartili interni)")
plt.show()

# DEMO 3 — Confronto fra categorie con hue

plt.figure(figsize=(10,6))
sns.violinplot(
    data=df, x="fascia", y="soddisfazione", hue="giorno",
    inner="stick", palette="Set3"
)
plt.title("Confronto tra fasce per giorno (violinplot)")
plt.legend(title="Giorno")
plt.show()

# DEMO 4 — Violin + box sovrapposti

plt.figure(figsize=(8,5))
sns.violinplot(data=df, x="giorno", y="soddisfazione", inner=None, cut=0)
sns.boxplot(
    data=df, x="giorno", y="soddisfazione",
    width=0.15, showcaps=True, boxprops={"facecolor":"none"}
)
plt.title("Boxplot sovrapposto a violinplot")
plt.show()
```
