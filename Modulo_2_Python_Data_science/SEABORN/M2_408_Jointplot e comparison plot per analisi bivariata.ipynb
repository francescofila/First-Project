```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

g = sns.jointplot(data=tips, x="total_bill", y="tip",
                  hue="sex", kind="kde",
                  fill=True, alpha=0.5)

g.plot_marginals(sns.kdeplot, linewidth=2)
g.plot_marginals(sns.rugplot, alpha=0.3, height=0.02)

plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

plt.figure(figsize=(9,5))
sns.violinplot(data=tips, x="sex", y="tip", hue="smoker", inner=None, split=True)
sns.boxplot(data=tips, x="sex", y="tip", hue="smoker", width=0.2)
sns.stripplot(data=tips, x="sex", y="tip", hue="smoker", dodge=True, alpha=0.6, size=3)

plt.title("Comparison: violin + box + punti (sex × smoker)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

g = sns.FacetGrid(tips, col="sex", row="smoker", hue="day", height=3.5)
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip", alpha=0.7)
g.map_dataframe(sns.kdeplot, x="total_bill", y="tip", levels=4, alpha=0.2)
g.add_legend()
plt.show()
```
